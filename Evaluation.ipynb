##3.3 Entraîner et ajuster les paramètres de plusieurs modèles

```python
#Entraîner au moins quatre modèles différents
models = {
    'DecisionTree': DecisionTreeClassifier(),
    'KNN': KNeighborsClassifier(),
    'RandomForest': RandomForestClassifier(),
    'SVM': SVC()
}

params = {
    'DecisionTree': {'max_depth': [3, 5, 7, 10]},
    'KNN': {'n_neighbors': [3, 5, 7, 9]},
    'RandomForest': {'n_estimators': [50, 100, 200]},
    'SVM': {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
}

#Et ajuster leurs hyperparamètres en utilisant GridSearchCV.
best_models = {}

for model_name in models:
    grid_search = GridSearchCV(models[model_name], params[model_name], cv=5, scoring='f1')
    grid_search.fit(X_train, y_train)
    best_models[model_name] = grid_search.best_estimator_
    print(f"Meilleurs paramètres pour {model_name}: {grid_search.best_params_}")
```

Les hyperparamètres permettent de trouver la combinaison optimale des paramètres qui minimise l'erreur et maximise les métriques de performance (le score F1, l'accuracy, etc.). ##3.4 Évaluer les performances des modèles

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
performance = {
    'Model': [],
    'Accuracy': [],
    'Recall': [],
    'F1 Score': []
}

for model_name, model in best_models.items():
    y_pred = model.predict(X_test)
    performance['Model'].append(model_name)
    performance['Accuracy'].append(accuracy_score(y_test, y_pred))
    performance['Recall'].append(recall_score(y_test, y_pred))
    performance['F1 Score'].append(f1_score(y_test, y_pred))

performance_df = pd.DataFrame(performance)

if not performance_df.empty:
    performance_df.set_index('Model').plot(kind='bar', figsize=(10, 6))
    plt.title("Comparison of Model Performance")
    plt.ylabel("Score")
    plt.show()
else:
    print("bon baa ça fonctione pas trop...")

#Aaahlala
```
